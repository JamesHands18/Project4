```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
diamond_df = pd.read_csv("./Resources/scaled.csv")
diamond_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = diamond_df['price'].values
y

X = diamond_df.drop('price', axis=1).values
X

# Split the preprocessed data into a training and testing dataset
X_train, X_test,y_train,y_test =train_test_split(X,y, random_state=10)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

diamonds_nn= len(X_train_scaled[0])
hidden_node_layer = 1
second_hidden_node_layer =100
third_hidden_layer= 150
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_node_layer, input_dim=diamonds_nn, activation='relu'))
#additional layer
nn.add(tf.keras.layers.Dense(units=third_hidden_layer, input_dim=diamonds_nn, activation='relu'))
# Second hidden layer
nn.add(tf.keras.layers.Dense(units=second_hidden_node_layer, activation='relu'))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='linear'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
trained_model = nn.fit(X_train_scaled, y_train, epochs=10)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
