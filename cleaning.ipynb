```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Resources/diamonds.csv')
```

```python
df
```

```python
df_dup = df.drop(columns = ['depth', 'table', "'x'", "'y'", "'z'"])
```

```python
dropable = df_dup[df_dup.duplicated()]
dropable
```

```python
drop_list = [i for i,row in dropable.iterrows()]
drop_list
```

```python
df.drop(drop_list, inplace = True)
df.reset_index(inplace=True, drop=True)
df.drop(columns=["'x'", "'y'", "'z'"], inplace=True)
df
```

```python
df.nunique()
```

```python
df = pd.get_dummies(df, columns=['cut', 'color', 'clarity'], drop_first=True)
```

```python
df.head(10)
```

```python
df.to_csv('Resources/cleaned.csv', index=False)
columns = df.columns.to_list()
```

```python
scaler = StandardScaler()
scaler_new = scaler.fit(df)
scaled_df = pd.DataFrame(scaler_new.transform(df), columns = columns)
```

```python
scaled_df.head(10)
```

```python
scaled_df.to_csv('Resources/scaled.csv', index=False)
```
